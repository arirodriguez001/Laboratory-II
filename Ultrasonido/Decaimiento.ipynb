```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
```

```python
gsheet = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vR_dTBWDtQGCgRD_ojD1YYlAQJMvm5EXJWpALyBhw2WAXwKQazmav_BB8BWBS8aMhnYCwjfkCiMZvw6/pub?output=csv'
data = pd.read_csv(gsheet, decimal='.')
x = data['Distancia (mm)'].to_numpy()
V = data['Voltaje (mV)'].to_numpy()
V_err = data['Escala (mV)'].to_numpy()/6
```

```python
plt.figure(figsize=(10,5))
plt.errorbar(x, V, yerr=V_err, fmt='.-', label="Datos")
#plt.xscale('log')
#plt.yscale('log')
plt.xlabel('Distancia (mm)')
plt.ylabel('Voltaje (mV)')


```

```python
def lineal(x, a, b):
  return a*x + b
```

```python
#V_out = V_in * r**alpha
#ln(V_out) = alpha*ln(r)+ln(V_in)
V_log = np.log(V)
x_log = np.log(x)
V_log_err = V_err/V
params, cov = curve_fit(lineal, x_log, V_log) #funcion para calcular los parámetros a partir de un ajuste
err = np.sqrt(np.diag(cov)) #funcion parte del curve para calcular errores
y_fit = lineal(x_log, params[0], params[1]) #lista de datos ajustados obtenidos por el curve fit
plt.figure()
plt.errorbar(x_log, V_log, yerr=V_log_err, fmt='.-', label = 'Datos')
plt.plot(x_log, y_fit, label = 'Ajuste')
plt.xlabel('Log(x)')
plt.ylabel('Log(V)')
plt.grid(), plt.legend()
print('La onda decae a razon de rˆ(', params[0], '+/-', err[0], ')')
```


*   log(V) = long(A0) + alfa*long(x)
*  log(V) = OAO + pendiente*long(x)



```python
print(f'La onda decae a razon de rˆ{params[0]:.2f} ± {err[0]:.2f} ')
```
