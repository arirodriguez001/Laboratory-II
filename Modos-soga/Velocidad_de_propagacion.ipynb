```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
```

```python
rho1 = 0.000042 #kg/m
rho1_err = 0.0000001 #kg/m
mT1 = 0.06369 #kg
mt1_err = 0.00001 #kg
L1 = 0.815 #m
L1_err = 0.01 #m
F1 = [76, 152, 229, 306] #Hz
F1_err = [2, 1, 2, 2] #Hz
l1 = [1.63, 0.815, 0.548, 0.408]
l1_err = [0.02, 0.02, 0.02, 0.03]

F2 = [30.5, 61, 92, 124]
F2_err = [0.5, 1, 1, 1]
l2 = [1.63, 0.82, 0.54, 0.41]
l2_err = [0.02, 0.02, 0.02, 0.03]

F3 = [21, 42, 64, 85, 107]
F3_err = [0.5, 1, 1, 2, 2]
l3 = [1.63, 0.82, 0.54, 0.41, 0.33]
l3_err = [0.02, 0.02, 0.02, 0.02, 0.03]
```

```python
def dispersion(F, v):
  return v/F
```

```python
params1, cov1 = curve_fit(dispersion, F1, l1, sigma = l1_err, absolute_sigma=True) #funcion para calcular los parámetros a partir de un ajuste
params2, cov2 = curve_fit(dispersion, F2, l2, sigma = l2_err, absolute_sigma=True)
params3, cov3 = curve_fit(dispersion, F3, l3, sigma = l3_err, absolute_sigma=True)

err1 = np.sqrt(np.diag(cov1)) #funcion parte del curve para calcular errores
err2 = np.sqrt(np.diag(cov2))
err3 = np.sqrt(np.diag(cov3))

y_fit1 = dispersion(F1, params1[0]) #lista de datos ajustados obtenidos por el curve fit
y_fit2 = dispersion(F2, params2[0])
y_fit3 = dispersion(F3, params3[0])

plt.figure()
plt.errorbar(F1, l1, yerr=l1_err, fmt='.')
plt.plot(F1, y_fit1, label = 'Tanza 1')

plt.errorbar(F2, l2, yerr=l2_err, fmt='.')
plt.plot(F2, y_fit2, label = 'Tanza 2')

plt.errorbar(F3, l3, yerr=l3_err, fmt='.')
plt.plot(F3, y_fit3, label = 'Tanza 3')

plt.ylabel('Longitud de onda [m]')
plt.xlabel('Frecuencia [Hz]')
plt.grid(), plt.legend()
print(f'La velocidad de propagación de la soga 1 es v1=({params1[0]:.1f} ± {err1[0]:.1f}) m/s')
print(f'La velocidad de propagación de la soga 2 es v2=({params2[0]:.1f} ± {err2[0]:.1f}) m/s')
print(f'La velocidad de propagación de la soga 3 es v3=({params3[0]:.1f} ± {err3[0]:.1f}) m/s')
```

```python
plt.figure()
plt.errorbar(F1, l1 - y_fit1, yerr=l1_err, fmt='.')
plt.plot(F1, y_fit1*0, label = 'Tanza 1')
plt.ylabel('Residuos [m]')
plt.xlabel('Frecuencia [Hz]')
plt.title('Residuos para la Tanza 1')
plt.figure()
plt.errorbar(F2, l2 - y_fit2, yerr=l2_err, fmt='.')
plt.plot(F2, y_fit2*0, label = 'Tanza 2')
plt.ylabel('Residuos [m]')
plt.xlabel('Frecuencia [Hz]')
plt.title('Residuos para la Tanza 2')
plt.figure()
plt.errorbar(F3, l3-y_fit3, yerr=l3_err, fmt='.')
plt.plot(F3, y_fit3*0, label = 'Tanza 3')
plt.ylabel('Residuos [m]')
plt.xlabel('Frecuencia [Hz]')
plt.title('Residuos para la Tanza 3')
```
